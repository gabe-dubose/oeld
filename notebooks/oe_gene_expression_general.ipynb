### Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import statistics
from gnat import rarefaction
```

### Data Loading

```python
oe_raw_counts = pd.read_csv('../data/oe_raw_counts_kallisto.csv', index_col=0)
```

```python
oe_raw_counts
```

#### Perform rarefaction

```python
samples = list(oe_raw_counts.index)
parameters = {'minimum_sample' : 1000, 'maximum_sample' : 150000, 'step' : 10000, 'data' : oe_raw_counts}
rare_curves = rarefaction.generate_rarefaction_curves(samples, parameters)
```

```python
sns.set_style("whitegrid")
fig, ax1 = plt.subplots(1, 1, figsize=(7, 5), sharey=True)

color_dict = {'3' : 'tab:blue',
             '5' : 'tab:olive',
             'E' : 'tab:green',
             'L' : 'tab:purple',
             'A' : 'tab:orange'}

for sample in rare_curves:
    sample_id = list(sample.keys())[0]
    stage = sample_id[5]
    values = sample[sample_id]['count']
    depth = sample[sample_id]['depth']
    sns.lineplot(y = values, x = depth, color=color_dict[stage])
    
# Creating custom legend
legend_labels = {
    '3': 'Third Instar',
    '5': 'Fifth Instar',
    'E': 'Early Pupa',
    'L': 'Late Pupa',
    'A': 'Adult'
}

# Adding legend to the plot
handles = [plt.Line2D([0], [0], color=color_dict[key], lw=2) for key in legend_labels]
labels = [legend_labels[key] for key in legend_labels]
legend = ax1.legend(handles, labels, title='Developmental Stage', loc='upper left', bbox_to_anchor=(1, 1))

ax1.set_ylabel('Number of genes detected', fontsize = 12)
ax1.set_xlabel('Sampling Depth', fontsize = 12)
plt.tight_layout()
plt.savefig('../figures/supplement/rarefaction.pdf')
```

#### Raw read counts

```python
#inspect read counts for each sample
oe_counts_total = oe_raw_counts.sum(axis=1)
oe_counts_total = pd.DataFrame(oe_counts_total, columns=['Total_Count'])
#add column for monarch stage
stage = []
plant = []
for sample in list(oe_counts_total.index):
    #add stage
    if sample[5] == '3':
        stage.append('third-instar')
    elif sample[5] == '5':
        stage.append('fifth-instar')
    elif sample[5] == 'E':
        stage.append('early-pupa')
    elif sample[5] == 'L':
        stage.append('late-pupa')
    elif sample[5] == 'A':
        stage.append('adult')
    
    #add plant
    if sample[6] == 'c':
        plant.append('A. curassavica')
    elif sample[6] == 'i':
        plant.append('A. incarnata')
             
oe_counts_total['stage'] = stage
oe_counts_total['plant'] = plant
oe_counts_total['sample_id'] = list(oe_counts_total.index)
oe_counts_total['log_counts'] = np.log(oe_counts_total['Total_Count'])

#sort data by transition order
stages = ['third-instar', 'fifth-instar', 'early-pupa', 'late-pupa', 'adult']
oe_counts_total['stage'] = pd.Categorical(oe_counts_total['stage'], categories = stages)
oe_counts_total = oe_counts_total.sort_values(by='stage')
```

```python
oe_counts_third = oe_counts_total.loc[oe_counts_total['stage'] == 'third-instar']
oe_counts_fifth = oe_counts_total.loc[oe_counts_total['stage'] == 'fifth-instar']
oe_counts_early = oe_counts_total.loc[oe_counts_total['stage'] == 'early-pupa']
oe_counts_late = oe_counts_total.loc[oe_counts_total['stage'] == 'late-pupa']
oe_counts_adult = oe_counts_total.loc[oe_counts_total['stage'] == 'adult']
```

```python
#plot read counts
sns.set_style("whitegrid")
fig, [ax1, ax2, ax3, ax4, ax5] = plt.subplots(1, 5, figsize=(15, 7), sharey=True)
sns.barplot(oe_counts_third, y='Total_Count', x="sample_id", ax=ax1, color="tab:gray")
sns.barplot(oe_counts_fifth, y='Total_Count', x="sample_id", ax=ax2, color="tab:gray")
sns.barplot(oe_counts_early, y='Total_Count', x="sample_id", ax=ax3, color="tab:gray")
sns.barplot(oe_counts_late, y='Total_Count', x="sample_id", ax=ax4, color="tab:gray")
sns.barplot(oe_counts_adult, y='Total_Count', x="sample_id", ax=ax5, color="tab:gray")

ax1.set_title('Third Instar', fontsize=15)
ax2.set_title('Fifth Instar', fontsize=15)
ax3.set_title('Early Pupa', fontsize=15)
ax4.set_title('Late Pupa', fontsize=15)
ax5.set_title('Adult', fontsize=15)

ax1.set_xticklabels([])
ax2.set_xticklabels([])
ax3.set_xticklabels([])
ax4.set_xticklabels([])
ax5.set_xticklabels([])

ax1.set_ylabel('Read Count', fontsize = 15)
ax2.set_ylabel('')
ax3.set_ylabel('')
ax4.set_ylabel('')
ax5.set_ylabel('')

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('')
ax4.set_xlabel('')
ax5.set_xlabel('')

ax1.tick_params(axis='y', labelsize=15)
plt.tight_layout()
plt.savefig('../figures/supplement/oe_read_counts.pdf')
```

### PCoA analysis using Manhattan distances

```python
#load data
manhattan_pca_data = pd.read_csv('../data/manhattan_oe_pcoa.csv', index_col=0)[["PC1", "PC2"]]
#add metadata data
stage = []
plant = []

for sample in list(manhattan_pca_data.index):
    if sample[5] == '3':
        stage.append('third-instar')
    elif sample[5] == '5':
        stage.append('fifth-instar')
    elif sample[5] == 'E':
        stage.append('early-pupa')
    elif sample[5] == 'L':
        stage.append('late-pupa')
    elif sample[5] == 'A':
        stage.append('adult')

    if sample[6] == 'i':
        plant.append('A. incarnata')
    elif sample[6] == 'c':
        plant.append('A. curassavica')
        
manhattan_pca_data['stage'] = stage
manhattan_pca_data['plant'] = plant
```

```python
#and now to plot
sns.set_style("whitegrid")
#define color palette
# Get the "Spectral" colormap
cmap = plt.cm.get_cmap('viridis')

# Divide the colormap into 14 bins
num_bins = 5
colors = [cmap(i / num_bins) for i in range(num_bins)]
# Extract hex codes
colors = [mcolors.to_hex(color) for color in colors]
#make palette
custom_palette = sns.color_palette(colors)
sns.set_palette(custom_palette)

fig, ax1 = plt.subplots(1, 1, figsize=(7.5, 5))
hue_order = ['third-instar', 'fifth-instar', 'early-pupa', 'late-pupa', 'adult']
manhattan_pca_plot = sns.scatterplot(data=manhattan_pca_data, x='PC1', y='PC2', ax=ax1, hue='stage', 
                style='stage', hue_order = hue_order,
                s=200, alpha=0.8, edgecolor = 'none', markers=['D', 's', 'h', 'p', 'o'])

ax1.set_xlabel('PCo1: 92.96%', size=20)
ax1.set_ylabel('PCo2: 3.29%', size=20)
ax1.set_xticklabels([])
ax1.set_yticklabels([])

handles, previous_labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles, 
           labels=['Third instar', 'Fifth instar', 'Early pupa', 'Late pupa', 'Adult'],
          bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=15, title=r'${D. plexippus}$ stage', 
           title_fontsize = 15, labelspacing = 1, markerscale=1.75)

plt.tight_layout()
plt.savefig('../figures/oe_transcriptomic_divergence.pdf')
```
