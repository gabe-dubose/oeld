### Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.lines as mlines
import numpy as np
```

### Quantifying transcriptional divergence betwen stages

Data was generated based on Manhattan distance calculated using the transcriptional\_turnover\_comparisons.R script.
    
OE

    Third to Fifth: 48,984.41
    Fifth to Early: 363,378.9
    Early to Late: 782,083.6
    Late to Adult: 806,003.9

Dplex

    Third to Fifth: 798,711.5
    Fifth to Early: 1,224,305
    Early to Late: 1,179,806
    Late to Adult: 930,267.8

```python
oe_transcriptional_turnover = [48984.41, 363378.9, 782083.6, 806003.9]
dplex_transcriptional_turnover = [798711.5, 1224305, 1179806, 930267.8]
transitions = ['3-5', '5-E', 'E-L', 'L-A']
```

```python
#initial plot
sns.set_style('whitegrid')

#plot
fig, ax1 = plt.subplots(1, 1, figsize=(5, 5))

#Dplex
sns.lineplot(y=dplex_transcriptional_turnover, x=transitions, ax=ax1, ls='--', color='tab:gray', linewidth=2.5)
#OE
sns.lineplot(y=oe_transcriptional_turnover, x=transitions, ax=ax1, linewidth=2.5, ls='-', color='black')

ax1.set_xlabel('Transition', fontsize = 15)
ax1.set_ylabel('Transcriptional Turnover', fontsize = 15)

ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)

#define legend
legend_lines = [mlines.Line2D([0], [0], color='tab:gray', label=r'${D. plexippus}$ (host)', linewidth=2, linestyle='--'),
                mlines.Line2D([0], [0], color='black', label=r'${O. elektroscirrha}$ (parasite)', linewidth=2)]

ax1.legend(handles=legend_lines, loc='lower right', fontsize=10)

plt.tight_layout()
```

##### Scaling turnover to gene count

Dplex has more genes and therefore more axes of variation. Since Manhattan distance represent the total transcriptional divergence, the fact that Dplex has more genes makes it more likely to have greater turnover due to simple addition. Therefore, I need to normalize the turnover based on gene counts.

D. plexippus gene count: 15233

OE gene count: 2634

```python
oe_gene_count = 2634
dplex_gene_count = 15233

oe_transcriptional_turnover_scaled = [value/oe_gene_count for value in oe_transcriptional_turnover]
dplex_transcriptional_turnover_scaled = [value/dplex_gene_count for value in dplex_transcriptional_turnover]
#oe_transcriptional_turnover_scaled = list(oe_transcriptional_turnover / oe_gene_count)
#dplex_transcriptional_turnover_scaled = list(dplex_transcriptional_turnover / dplex_gene_count)
```

```python
#scaled plot
sns.set_style('whitegrid')

#plot
fig, ax1 = plt.subplots(1, 1, figsize=(5, 5))

#Dplex
sns.lineplot(y=dplex_transcriptional_turnover_scaled, x=transitions, ax=ax1, ls='--', color='tab:gray', linewidth=2.5)
#OE
sns.lineplot(y=oe_transcriptional_turnover_scaled, x=transitions, ax=ax1, linewidth=2.5, ls='-', color='black')

ax1.set_xlabel('Transition', fontsize = 15)
ax1.set_ylabel('Normalized Transcriptional Turnover', fontsize = 15)

ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)

#define legend
legend_lines = [mlines.Line2D([0], [0], color='tab:gray', label=r'${D. plexippus}$ (host)', linewidth=2, linestyle='--'),
                mlines.Line2D([0], [0], color='black', label=r'${O. elektroscirrha}$ (parasite)', linewidth=2)]

ax1.legend(handles=legend_lines, loc='lower right', fontsize=10)

plt.tight_layout()
```

### Distance correlations

```python
#load data
spearman_data = pd.read_csv('../data/transcriptional_correlation_dplex_oe_spearman.csv')
pearson_data = pd.read_csv('../data/transcriptional_correlation_dplex_oe_pearson.csv')

#remove diagnonals
pearson_data = pearson_data.loc[(pearson_data['oe.pearson.vector'] != 0) & (pearson_data['dplex.pearson.vector'] != 0)]
spearman_data = spearman_data.loc[(spearman_data['oe.spearman.vector'] != 0) & (spearman_data['dplex.spearman.vector'] != 0)]
```

##### Pearson

Mantel test:

    Mantel statistic r: 0.8132 
          Significance: 0.001 
    
    Upper quantiles of permutations (null model):
       90%    95%  97.5%    99% 
    0.0455 0.0601 0.0785 0.0889 
    Permutation: free
    Number of permutations: 999

```python
#plot
sns.set_style('whitegrid')

#plot
fig, ax1 = plt.subplots(1, 1, figsize=(5, 5))

sns.regplot(y='oe.pearson.vector', x='dplex.pearson.vector', data=pearson_data, ax=ax1, scatter=False, color='black')
sns.scatterplot(y='oe.pearson.vector', x='dplex.pearson.vector', data=pearson_data, ax=ax1, color='black', edgecolor='none', alpha=0.25)

ax1.set_xlabel('$D. plexippus$ transcriptional turnover', fontsize = 15)
ax1.set_ylabel('$O. elektroscirrha$ transcriptional turnover', fontsize = 15)

ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)
```
